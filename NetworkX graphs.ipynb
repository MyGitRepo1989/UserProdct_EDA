```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("networkgraph.csv")
```

```python
df.head(5)
```

```python
df=df[['antecedents','consequents','support']]
```

```python
a1= [x.split("{") for x in df['antecedents']]
```

```python
a2=[x[1].split('}')[0] for x in a1]
```

```python
df['antecedents']=pd.Series(a2)
```

```python
a1= [x.split("{") for x in df['consequents']]
```

```python
a2=[x[1].split('}')[0] for x in a1]
```

```python
df['consequents']=pd.Series(a2)
```

```python
df1=df.copy()
```

```python
df1.columns=['Source','Target','weight']
```

```python
import networkx as nx
```

```python
G = nx.from_pandas_edgelist(df1, 
                            source='Source', 
                            target='Target', 
                            edge_attr='weight')
```

```python
print("No of unique characters:", len(G.nodes))
print("No of connections:", len(G.edges))
```

```python
graphs_viz_options = [nx.draw, nx.draw_networkx, nx.draw_circular, nx.draw_kamada_kawai, nx.draw_random, nx.draw_shell, nx.draw_spring]
```

```python
selected_graph_option = 5
```

```python
# plot
plt.figure(figsize=(8,6), dpi=100) 
nx.draw(G,with_labels=True,font_weight='normal',font_size=7,font_color='black')

#graphs_viz_options[selected_graph_option](G)
```

```python
import random
random.seed(246)        # or any integer
import numpy
numpy.random.seed(4812)
```

```python
from networkx import path_graph, random_layout
pos = random_layout(G, seed=None)  # use (either) global default RNG
pos = random_layout(G, seed=42)  # local RNG just for this call
pos = random_layout(G, seed=numpy.random)  # use numpy global RNG
random_state = numpy.random.RandomState(42)
pos = random_layout(G, seed=random_state)  # use/reuse your own RNG
```

```python
plt.figure(figsize=(17,12))
nx.draw_networkx(G,with_labels=True,node_color='#fca7f4',node_size=1200)
plt.grid(alpha=.5)
plt.show()
```
