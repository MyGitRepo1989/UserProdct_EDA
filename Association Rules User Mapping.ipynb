<h2> Most Popular Screens - Support

<h3> What is support?</h3><br>
<b>This says how POPULAR an itemset is </b>, as measured by the proportion of transactions in which an itemset appears.
<br><br>
 MOST POPULAR SCREENS - How many times this Screen will be used compared to others?<br>

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_iemsets = apriori(df, min_support=0.2, use_colnames =True)
frequent_iemsets.sort_values(by=['support'],ascending=False)[:15]
```

<H1> Next Screen Used ? - confidence

Confidence - This says how likely Product B (consequents) is purchased when Product A (antecedents) is purchased, expressed as {A -> B}. <br>This is measured by the proportion of transactions with Product A, in which Product B also appears. 

```python
from mlxtend.frequent_patterns import association_rules
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
res = association_rules(frequent_iemsets,metric='confidence',min_threshold=0.7)
```

```python
res.columns
```

<br><h3>What is 'antecedents', 'consequents' ? </h3> <br>
<i>The IF component of an association rule is known as the antecedent. The THEN component is known as the consequent. The antecedent and the consequent are disjoint; they have no items in common.

```python
res[['antecedents', 'consequents',  'support', 'confidence', 'lift']][:15]
```

```python
res1=res[['antecedents', 'consequents',  'support', 'confidence', 'lift']]
```

```python
res1.columns=['FIRST SCREEN', 'THEN NEXT SCREEN',  'support', 'confidence', 'lift']
```

<br><br>
<b> CONFIDENCE</b> = Measure of how strongly the First Screens and Next Screens go together<br>
<b> SUPPORT</b> = How POPULAR the INDIVIDUAL Group of Screens are<br>
<b> LIFT </b> = How LIKELY is NEXT SCREEN going to come if First Screen is clicked<br>
<br><br>

```python
res1[res1['confidence']>=1].sort_values(by=['support'],ascending=False)[:25]
```

<br>
<h2> What is Lift?</h2><br>
Lift basically tells us that the likelihood of buying a Product A and Product B together is 3.33 times more than the likelihood of just buying the Product B. <br><br>A Lift of 1 means there is no association between products A and B. <br><br>Lift of greater than 1 means products A and B are more likely to be bought together<br><br>

```python
res1[res1['support']>=.5].sort_values(by=['lift'],ascending=False)[:20]
```
